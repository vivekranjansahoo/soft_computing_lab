1. 	WAP to plot the membership values of elements U={A,B,C,D,E,F,G,H,I,J,K,M,N} as
 F={ (A,0.1),(B,0.9),(C,0.2),(D,0.4),(E,0.6),(F,0.3),(G,0.4),(H,0.6),(I,0.2),(J,0),(K,0.1),(M,0.6),(N,0.5)}

```python
import matplotlib.pyplot as plt
import numpy as np

# Membership function
def membership_function(x, theta=10):
    return 1 if x > theta else 0

# Define the continuous range of X
X = np.linspace(-20, 20, 1000)
F = [membership_function(x) for x in X]

# Plotting the membership values
plt.figure(figsize=(10, 6))
plt.step(X, F, color='skyblue', linewidth=2, where='post')

# Adding labels and title
plt.xlabel('Elements of X', fontsize=12)
plt.ylabel('Membership Values (f(x))', fontsize=12)
plt.title('Membership Function Plot with Threshold \u03B8=10', fontsize=14)
plt.ylim(-0.1, 1.1)

# Show grid for better visualization
plt.grid(axis='both', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

 11.	WAP to enter a discrete unorder Universal set X. Built a fuzzy set A on it and plot the graph.

```python
import matplotlib.pyplot as plt

# Step 1: Input Universal Set X (discrete unordered)
X = list(map(int, input("Enter the elements of the universal set X (separated by spaces): ").split()))

# Step 2: Build the fuzzy set A
A = {}
print("Enter the membership values for the fuzzy set A (between 0 and 1):")
for x in X:
    while True:
        try:
            membership_value = float(input(f"Membership value for {x}: "))
            if 0 <= membership_value <= 1:
                A[x] = membership_value
                break
            else:
                print("Please enter a value between 0 and 1.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")

# Step 3: Plot the graph
elements = list(A.keys())
membership_values = list(A.values())

plt.figure(figsize=(8, 6))
plt.plot(elements, membership_values, marker='o', linestyle='-', color='b')
plt.title("Fuzzy Set A on Universal Set X")
plt.xlabel("Elements of X")
plt.ylabel("Membership Values of A")
plt.grid(True)
plt.show()
```

13.	WAP to enter a continuous universal set X. Built a fuzzy set A on X and plot the graph.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Input the range for the continuous Universal Set X
start = float(input("Enter the start value of the continuous set X: "))
end = float(input("Enter the end value of the continuous set X: "))
step = float(input("Enter the step size for the continuous set X: "))

# Generate the continuous set X
X = np.arange(start, end, step)

# Step 2: Build the fuzzy set A (membership function)
# Define a simple fuzzy membership function (for example, a linear function)
def fuzzy_membership(x):
    return np.clip(1 - abs(x - (start + end) / 2) / ((end - start) / 2), 0, 1)

# Apply the fuzzy membership function to X
A = fuzzy_membership(X)

# Step 3: Plot the graph
plt.figure(figsize=(8, 6))
plt.plot(X, A, marker='o', linestyle='-', color='b')
plt.title("Fuzzy Set A on Continuous Universal Set X")
plt.xlabel("Elements of X")
plt.ylabel("Membership Values of A")
plt.grid(True)
plt.show()
```
